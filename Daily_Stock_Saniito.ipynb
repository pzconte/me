#Inventario Walmart

```python
import pandas as pd
import warnings
from ipywidgets import interact, widgets
import datetime as dt

warnings.simplefilter("ignore")

date = '04-19-2023'
path_stock = f'/content/drive/MyDrive/Saniito/a) Data B2B/Walmart/csv/Stock {date}.xlsx'
stock_walmart = pd.read_excel(path_stock, header=17, engine="openpyxl")
stock_walmart.drop(['Legacy PLU Nbr', 'Item Nbr', 'Item Desc 2',
                    'UPC',	'UPC Desc',	'Store Nbr', 'Region'], axis=1, inplace=True)

print("*"*12 + " STOCK SANIITO B2B " + "*"*12 + "\n")
col_loc = 'Store Name'
loc_name = "EX PLYON"
col_stock = 'Curr Str On Hand Qty'
col_product = 'Item Desc 1'
col_address = "Building Address"
col_city= "City"
df = stock_walmart.sort_values(by="Store Name")
def update_names(loc_name):
    quantity = df.loc[df[col_loc]==loc_name, col_stock].unique()
    local = df.loc[df[col_loc]==loc_name, col_product].unique()
    stock = dict(zip(local, quantity))
    address = df.loc[df[col_loc]==loc_name, col_address].unique()[0]
    city = df.loc[df[col_loc]==loc_name, col_city].unique()[0]
    for key in stock:
        print(f'{key}: {stock[key]}')
    print(f"Dirección: {address} - {city}")
    print("-"*44)
    return f'Información actualizada al {dt.datetime.strptime(date, "%m-%d-%Y").strftime("%d/%m/%Y")}'

names_dropdown = widgets.Select(options=df[col_loc].unique(), description='Local:')

widgets.interact(update_names, loc_name=names_dropdown)
print("--")
```
